```python
file_path = "Files/Superstore.csv"
df = spark.read.option("header", "true").csv(file_path)

print("Original Columns:", df.columns)
for col_name in df.columns:
    df = df.withColumnRenamed(col_name, col_name.replace(" ", "_"))

print("Renamed Columns:", df.columns)

df.write.mode("overwrite").format("delta").saveAsTable("SalesData")

df2 = spark.read.table("SalesData")
df2.show(5)

df_clean = df2.dropna()

from pyspark.sql.functions import to_date, col

df_clean = df_clean.withColumn("Order_Date", to_date(col("Order_Date"), "MM/dd/yyyy"))

#Add new column Total_Profit = Profit * 1.05
df_clean=df_clean.withColumn("Total_profit",col("Profit")* 1.05)

spark.conf.set("spark.sql.legacy.timeParserPolicy","LEGACY")

df_clean.show(5)

df_clean.write.mode("overwrite").format("delta").saveAsTable("CleanSalesData")
```
